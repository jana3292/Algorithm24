## 1. 문제정의

카운팅 정렬

## 2. 알고리즘 설명

입력의 종류에 따라서는 리스트의 각 항목들을 단순히 세는 방법으로 정렬할 수 있다. 이런 정렬 기법을 카운팅 정렬이라고 한다.

## 4. 코드개요

함수명: counting_sort

입력:
A: 정렬할 리스트
변환값: 없음

1. output: 정렬 결과를 저장할 임시 리스트를 생성합니다.
2. count: 각 숫자의 빈도를 저장할 리스트를 생성합니다.
3. 입력 리스트 A를 순회하며 각 숫자별 빈도를 계산하여 count 리스트에 저장합니다.
4. count 리스트를 수정하여 각 숫자의 위치가 되도록 조정합니다.
5. 입력 리스트 A를 다시 순회하며, 각 숫자를 정렬 결과 리스트 output의 적절한 위치에 저장합니다.
6. 정렬 결과를 원래 배열 A에 복사합니다.


## 5. 코드

```python
def counting_sort(A): 
    output = [0] * len(A) # 정렬 결과저장용 임시 리스트 
    count = [0] * MAX_VAL # 각 숫자의 빈도를 저장 
    for i in A: # 각 숫자별 빈도를 계산 
        count[i] += 1 
    for i in range(MAX_VAL): # count[i]가 출력 배열에서 
        count[i] += count[i-1] # 해당 숫자의 위치가 되도록 수정
    for i in range(len(A)): # 모든 입력항목 A[i]에 대해 
        output[count[A[i]]-1] = A[i] # 해당 위치(count[A[i]]-1)에 저장 
        count[A[i]] -= 1 # 킷값 A[i]의 위치를 하나 줄임 
    for i in range(len(A)): # 정렬 결과를 원래 배열에 복사 
        A[i] = output[i]
```

## 6. 테스트 코드

```python
def counting_sort(A): 
    output = [0] * len(A) # 정렬 결과저장용 임시 리스트 
    count = [0] * MAX_VAL # 각 숫자의 빈도를 저장 
    for i in A: # 각 숫자별 빈도를 계산 
        count[i] += 1 
    for i in range(MAX_VAL): # count[i]가 출력 배열에서 
        count[i] += count[i-1] # 해당 숫자의 위치가 되도록 수정
    for i in range(len(A)): # 모든 입력항목 A[i]에 대해 
        output[count[A[i]]-1] = A[i] # 해당 위치(count[A[i]]-1)에 저장 
        count[A[i]] -= 1 # 킷값 A[i]의 위치를 하나 줄임 
    for i in range(len(A)): # 정렬 결과를 원래 배열에 복사 
        A[i] = output[i]

MAX_VAL = 10
data = [1, 4, 1, 2, 7, 5, 2]
print("Original : ", data)
counting_sort(data)      #카운팅 정렬
print("Counting : ", data)
```

## 8. 복잡도 분석

카운팅 정렬의 시간 복잡도는 입력 데이터의 크기 
n과 입력의 범위 k에 따라 결정됩니다. 만약 입력의 크기가 𝑛
n이고, 숫자의 범위가 k라면 전체 알고리즘은 
O(k+n)의 시간 복잡도를 가지고있습니다.
